# 1- some infectious varient and there month wise analysis

```python
df['variant'].unique()
```

```python
plt.figure(figsize = (10, 10))
df.loc[df['variant'] == 'Beta'].groupby('month')['num_sequences'].agg('sum').plot()
plt.title("BETA varient")
```

1**.By this graph we can see, BETA cases start increasing from starting of year and start decreasinf from may and again start increasing from october. **

```python
plt.figure(figsize = (10, 10))
df.loc[df['variant'] == 'Omicron'].groupby('month')['num_sequences'].agg('sum').plot()
plt.title("omicron varient")
```

1.**by this graph we can see, omicron cases start increasing from october and at peak at november and december.**

```python
plt.figure(figsize = (10, 10))
df.loc[df['variant'] == 'Kappa'].groupby('month')['num_sequences'].agg('sum').plot()
plt.title("Kappa varient")
```

1.**by this graph we can see, kappa  cases start increasing from febraury and at peak in may  and start decreaasing from june.**

```python
plt.figure(figsize = (10, 10))
df.loc[df['variant'] == 'Gamma'].groupby('month')['num_sequences'].agg('sum').plot()
plt.title("Gamma varient") 
```

**1.by this graph we can see, gamma cases start increasing from march and at peak at june and start decreasing from july.**

```python
plt.figure(figsize = (10, 10))
df.loc[df['variant'] == 'Alpha'].groupby('month')['num_sequences'].agg('sum').plot()
plt.title("Alpha varient")
```

**this case also increas two times in a year like beta varient ,by this graph can see cases start increasing from strating of year and again start from october.**


# 2- variant cases saperately by location and date .

```python
df.head(5)
```

```python
columns = df.columns
columns
```

```python
df['month'] = pd.DatetimeIndex(df['date']).month
df['month']
```

```python
df['date_new']=pd.to_datetime(df['date'])
```

```python
df_variant=df['date_new'].dt.month.value_counts().to_frame()
df_variant.style.background_gradient(cmap='Blues')
```

```python
df['date_new'].dt.month.value_counts().plot(kind='bar')
plt.rcParams["figure.figsize"] = (10,10)
```

**this graoh showing month wise cases .**

```python
last_df=df.groupby(["location","variant"])['num_sequences'].sum()
```

```python
last_dfs=last_df.sort_values(ascending=False).head(30).to_frame()
last_dfs.style.background_gradient(cmap='Blues')
```

**here we can see location and variant wise case**s.

```python
data_df = df.groupby(['date_new'])['location','num_sequences_total','variant'].max()
```

```python
data_df.sort_values(by=["num_sequences_total",'variant'],ascending=False).head(50)
```
